```python
import pandas as pd
import zipfile
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
with zipfile.ZipFile('sales_train.csv.zip', 'r') as zip_ref:
    zip_ref.extractall('data')

df = pd.read_csv('data/sales_train.csv')
print(df.head())
```

```python
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y')

sales_by_date = df.groupby('date')['item_price'].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(sales_by_date['date'], sales_by_date['item_price'], label='Price', color='red')
plt.title('price Trend Over Time')
plt.xlabel('Date')
plt.ylabel('price')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
def create_lagged_features(df, lag=1):
    lagged_data = df.copy()
    for i in range(1, lag+1):
        lagged_data[f'lag_{i}'] = lagged_data['item_price'].shift(i)
    return lagged_data

lag = 5
sales_with_lags = create_lagged_features(df[['date', 'item_price']], lag)

sales_with_lags = sales_with_lags.dropna()
```

```python
X = sales_with_lags.drop(columns=['date', 'item_price'])
y = sales_with_lags['item_price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)
```

```python
model_xgb = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1, max_depth=5)
model_xgb.fit(X_train, y_train)
```

```python
predictions_xgb = model_xgb.predict(X_test)
rmse_xgb = np.sqrt(mean_squared_error(y_test, predictions_xgb))

print(f"RMSE: {rmse_xgb:.2f}")
```

```python
plt.figure(figsize=(12, 6))
plt.plot(y_test.index, y_test, label='Actual price', color='red')
plt.plot(y_test.index, predictions_xgb, label='Predicted price', color='green')
plt.title('Sales Forecasting using XGBoost')
plt.xlabel('Date')
plt.ylabel('price')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
